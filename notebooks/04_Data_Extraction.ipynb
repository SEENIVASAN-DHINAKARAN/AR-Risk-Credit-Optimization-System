```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:SEEni2003%40@localhost:1611/ar_risk_system")

# Load tables
customers_df = pd.read_sql("SELECT * FROM customers", engine)
invoices_df = pd.read_sql("SELECT * FROM invoices", engine)
payments_df = pd.read_sql("SELECT * FROM payments", engine)
customer_features_df = pd.read_sql("SELECT * FROM customer_features", engine)
customer_risk_output_df = pd.read_sql("SELECT * FROM customer_risk_output", engine)

print("All tables loaded successfully!")
```

```python
file_path = "../data_backup/AI_AR_Complete_Output.xlsx"

with pd.ExcelWriter(file_path, engine="openpyxl") as writer:
    customers_df.to_excel(writer, sheet_name="Customers", index=False)
    invoices_df.to_excel(writer, sheet_name="Invoices", index=False)
    payments_df.to_excel(writer, sheet_name="Payments", index=False)
    customer_features_df.to_excel(writer, sheet_name="Customer_Features", index=False)
    customer_risk_output_df.to_excel(writer, sheet_name="Customer_Risk_Output", index=False)

print("Complete Excel file created successfully!")
```
